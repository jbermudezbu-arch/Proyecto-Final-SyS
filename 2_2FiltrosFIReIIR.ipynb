# ¿Que son los filtros FIR?

FIR o Finite Impulse Response. Si le metes un impulso al filtro, la salida dura un número finito de muestras y luego se vuelve cero.

Matematicamente, un FIR es una suma ponderada de muestras pasadas sin realimentación.

El FIR es un promediador generalizado, asi que cada muestra de salida es una mezcla controlada del pasado y el presente.

### Propiedades:

Su estabilidad es garantizada

Si los coeficientes son simetricos el FIR no distorsiona la forma de la señal y solo introduce un retardo constante

### Desventaja principal:

Le cuesta lograr transiciones bruscas en frecuencia, para ello necesita muchos coeficientes, tiene mayor costo computacional y más retardo.



# ¿Que son los filtros IIR?

IIR o Infinite Impulse Response. Contrario a los FIR si a este filtro se mete un impulso, la salida no se acaba nunca, idealmente decaera con el tiempo pero no llegara a cero.

Fisicamente un IIR es como un sistema masa resorte amortiguador discreto con realimentación.


### Ventaja principal:

Altamente selectivo con bajo orden, logra pendientes fuertes con pocos coeficientes, esto hace que el retardo sea pequeño

### Desventajas principales:

Puede ser inestable si sus polos salen del circulo unitario, su Fase es no lineal asi que distorsiona la forma de onda.

# ¿Como se utilizan para resaltar patrones espaciales, temporales y espectrales desde señales EEG en tareas de MI?

En MI el EEG cambia en espacio pues son diferentes las zonas motoras de la izquierda y de la derecha, en el tiempo pues los cambios se dan en ventanas especificas y en espectro pues hay subidas y caidas de potencia en diferentes bandas.

El rol de los filtros es filtrar el EEG para aislar bandas relevantes haciendo uso de FIR pasa banda o IIR Butterworth en caso de haber restricciones de tiempo. El FIR es útil aqui pues no distorsiona la señal.

Luego se pueden usar FIR o IIR para resaltar patrones temporales. Como la MI no ocurre todo el tiempo se divide el EEG en ventanas temporales y dependiendo: si se requiere estabilidad se usa un FIR largo que suaviza el ruido pero tiene menor resolución temporal o si se requiere una respuesta más rápida y menos retardo se usa un IIR

Finalmente para resaltar patrones espaciales se usan los filtros FIR e IIR para asegurar que la varianza venga de MI y no de parpadeos, EMG o DC drift, para luego usar CSP con el fin de extraer patrones temporales.
